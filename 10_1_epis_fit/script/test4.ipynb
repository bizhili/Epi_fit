```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture

# 生成样本数据
np.random.seed(42)
num_samples = 1000
samples = np.concatenate([
    np.random.normal(loc=2, scale=1, size=num_samples // 2),
    np.random.normal(loc=6, scale=0.5, size=num_samples // 2)
])

# 拟合混合高斯模型
gmm = GaussianMixture(n_components=2)
gmm.fit(samples.reshape(-1, 1))

# 输出拟合结果
print("Gaussian Means:", gmm.means_.flatten())
print("Gaussian Standard Deviations:", np.sqrt(gmm.covariances_).flatten())
print("Gaussian Weights:", gmm.weights_.flatten())

# 可视化拟合结果
x = np.linspace(-10, 10, 1000)
pdf = np.exp(gmm.score_samples(x.reshape(-1, 1)))
plt.hist(samples, bins=30, density=True, alpha=0.5, label='Data')
plt.plot(x, pdf, '-r', label='GMM Fit')
plt.legend()
plt.show()
```
