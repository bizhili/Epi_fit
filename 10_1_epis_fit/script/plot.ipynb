```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoidDif(x):
    return sigmoid(x)*(1-sigmoid(x))

def certainty(a):
    return 1-2/(np.exp(0.5*a)+1)

# Generate x values from -10 to 10
x = np.linspace(0, 100, 10000)

# Calculate the corresponding y values using the sigmoid function
y = certainty(x)


# Plot the sigmoid function graph
plt.plot(x, y, label='certainty')
plt.xlabel('a')
plt.ylabel('Certainty')
plt.title('Certainty')
plt.legend()
plt.grid(True)
plt.axhline(y=0, color='gray', linestyle='dashed')
plt.axvline(x=0, color='gray', linestyle='dashed')
plt.xscale("log")
plt.show()
```
