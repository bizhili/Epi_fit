```python
import numpy as np
import matplotlib.pyplot as plt

mean = 100  # Replace this with your desired mean
variance = 4  # Replace this with your desired variance

upper= int(mean + 3 * np.sqrt(variance))
lower= int(mean - 3 * np.sqrt(variance))
lower= lower if lower>0 else 0
x_values = np.arange(lower, upper) 
probabilities = (1 / np.sqrt(2 * np.pi * variance)) * np.exp(-(x_values - mean) ** 2 / (2 * variance))


plt.plot(x_values, probabilities, label='Normal Distribution')
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.title('Normal Distribution with mean = {} and variance = {}'.format(mean, variance))
plt.legend()
plt.grid(True)
plt.show()
```
